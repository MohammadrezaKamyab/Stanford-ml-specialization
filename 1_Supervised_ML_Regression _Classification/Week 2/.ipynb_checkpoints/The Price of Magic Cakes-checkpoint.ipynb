```python
height = np.random.uniform(5, 15, number_cakes) # height : 5 - 15 cm
diameter = np.random.uniform(10,30,number_cakes) # diameter : 10 - 30 cm
dragon_sugar_grams = np.random.uniform(100,500, number_cakes) # dragon sugar : 100 - 500 g

# Price = 3 * height + 2 * diameter + 0.5 dragon_sugar_grams + noise
noise = np.random.normal(0,10,number_cakes)
price = 3 * height + 2 * diameter + 0.5 * dragon_sugar_grams + noise

cake_df = pd.DataFrame({
    "height" : height,
    "diameter" : diameter,
    "dragon_sugar_grams" : dragon_sugar_grams,
    "price" : price}
)
cake_df.head()
```

```python
# Preparing Data for Training
x = cake_df.iloc[:, :-1]
y = cake_df.iloc[:, -1]
```

<h2>üî¨ Chapter 4: Building the Spell </h2>
With quill and logic, you began to craft the prediction spell. Each ingredient became a term in your incantation.

You whispered to your scroll:

‚ÄúThe price is not a mystery... it is a sum of weighted effects. Height has its impact. So does diameter. And dragon sugar. Their weights‚Äîcoefficients‚Äîmust be learned.‚Äù

For days, you tested your spell using the old data. You adjusted the coefficients until they made accurate predictions. When you tested the spell on known cakes, the predictions were nearly perfect!



```python
def compute_output(x, w, b):
    """
    args:
        x (ndarray , shape(m,n)) : feature variable with n features and m examples
        w (ndarray , shape(n)) : weights of each feature variables
        b (scalar) : bias
    return:
        f_wb (m) : predictions
    """
    return np.dot(x, w) + b
```

```python
def compute_cost (x, y, w, b):
    """
    args:
        x (ndarray , shape(m,n)) : feature variable with n features and m examples
        y (ndarray , shape(m)) : target values
        w (ndarray , shape(n)) : weights of each feature variables
        b (scalar) : bias
    return:
        J_wb (scalar) : How well our model is doing, the lower the J the better the model
    """
    m = y.shape[0]
    f_wb = compute_output(x,w,b)
    J_wb = (1 / 2 * m) * np.sum(np.square(f_wb - y))
    return J_wb
```

```python
def compute_gradient (x, y, w, b):
    """
    args:
        x (ndarray , shape(m,n)) : feature variable with n features and m examples
        y (ndarray , shape(m)) : target values
        w (ndarray , shape(n)) : weights of each feature variables
        b (scalar) : bias
    return:
       dj_dw (ndarray , (n,)) : gradient of cost w.r.t. w
       dj_db (scalar) : gradient of cost w.r.t. b
    """
    m = y.shape[0]
    f_wb = compute_output(x,w,b)
    dj_dw = np.dot(x.T, f_wb - y) * (1/m)
    dj_db = np.sum(f_wb - y) * (1/m)
    return dj_dw, dj_db
```

```python
def gradient_descent(x, y, w_in, b_in, alpha, num_iter, cost_function, gradient_function, tolerance=1e-6):
    """
    Args:
        x (ndarray (m,n)) Feature variable with m examples and n features
        y (ndarray (m,)) Target variable
        w_in (ndarray , (n,)) : initialization of vector w
        b_in (scalar) : initialization b
        alpha (float) : learning rate
        num_iter (int): number of iteration for gradient descent
        cost_function : squared error cost function
        gradient function : generate dj_dw, dj_db
        tolerance : when to stop the gradient descent
    Returns:
        w (ndarray (n,)) : weights after the gradient descent
        b (scalar): b after the gradient descent 
        J_history (list) : list of cost function history
    """
    w = copy.deepcopy(w_in)
    b = b_in
    J_history = list()

    for i in range(num_iter):
        dj_dw , dj_db = compute_gradient(x, y, w , b)
        w = w - alpha * dj_dw
        b = b - alpha * dj_db

        if (i < 100000):
            J_history.append(cost_function(x,y,w,b))

        if (i % math.ceil(num_iter / 10) == 0):
            print(f"Iteration {i:5d} : Cost = {J_history[-1]:8.4f}")

        if (i > 0 and abs(J_history[-1] - J_history[-2]) < tolerance):
            print(f"Converges at index {i}")
            break
    return w, b, J_history
```

```python
w_init = np.zeros(x.shape[1])
b_init = 0
alpha = 0.00001
iterations = 100000

w_final, b_final , J_hist = gradient_descent(x, y, w_init, b_init, alpha, iterations, compute_cost, compute_gradient)

plt.plot(J_hist)
plt.title("Learning Curve")
plt.xlabel("Iteration")
plt.ylabel("Cost")
plt.grid(axis='both', lw=0.5, alpha=0.5 , linestyle='--')
```

```python
print(f"W = {w_final}\nb = {b_final}")
```
