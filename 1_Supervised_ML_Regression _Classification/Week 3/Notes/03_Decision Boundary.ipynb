```python
def sigmoid (z):
    """
    Compute sigmoid of Z
    Args:
        z (ndarray): shape(m) m training example
    returns:
        g (ndarray): same shape as z , sigmoid of z
    """
    return 1 / (1 + np.exp(-z))
```

```python
def compute_output(x,w,b):
    """
    Compute output for Logistic regression Model
    Args:
        x (ndarray): shape(m, n) , m is the number of training example and n is the number of features
        w (ndarray): shape(n), weights for each feature
        b (scalar): bias, parameter
    returns:
        prediction_y (ndarray): shape(m)
    """
    z = np.dot(x, w) + b
    return sigmoid(z)
```

## Is the value of Y going to be **0** or **1**
* One way is to set a **threshold** where
* if f_wb(x) < <strong>threshold</strong> => y_hat = 0
* if f_wb(x) >= **threshold** => y_hat = 1

* a common choice is to pick a **threshold = 0.5**
* if f_wb(x) < 0.5 => y_hat = 0
* if f_wb(x) >= 0.5 => y_hat = 1

- When y_hat = 1 ? **when f_wb(x) >= 0.5**
- When f_wb(x) >= 0.5 ? **When g(z) >= 0.5**
- When g(z) >= 0.5 ? **When z >= 0**

- When z >= 0 ? **When (x.w + b) >= 0**


<strong>y_hat = 1 when (w.x + b) >= 0</strong>
